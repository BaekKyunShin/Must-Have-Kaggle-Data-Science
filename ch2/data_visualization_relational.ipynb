## 관계도

```python
import seaborn as sns

flights = sns.load_dataset('flights') # 비행기 탐승자 수 데이터 불러오기
```

```python
flights.head()
```

### 히트맵(heatmap)

```python
flights_pivot = flights.pivot(index='month', 
                              columns='year', 
                              values='passengers')

flights_pivot
```

```python
sns.heatmap(data=flights_pivot);
```
